```python
#Milestone 2 Task 4

import pandas as pd
import seaborn as sns

def load_and_process(url):
    data = pd.read_csv(url)
    df = pd.DataFrame(data)
    return df

url = 'https://raw.githubusercontent.com/data301-2020-winter1/course-project-group_118/main/data/raw/plays.csv'
df = load_and_process(url)

##### columns : "play_id","game_id","play_num","team_id_for","team_id_against","event","secondaryType","x","y","period","periodType","periodTime","periodTimeRemaining","dateTime","goals_away","goals_home","description","st_x","st_y","rink_side"

df2 = df.drop(['game_id', 'team_id_for', 'team_id_against', 'secondaryType', 'x', 'y', 'period', 'periodType', 'periodTimeRemaining', 'dateTime', 'goals_away', 'goals_home', 'description', 'st_x', 'st_y', 'rink_side'], axis = 1)

df2 = df2[df.event.isin(['Faceoff', 'Goal', 'Stoppage', 'Period Start', 'Period End'])]

counts = df2.groupby(['event'], as_index = False).count()
#mydf2['total_goals'] = mydf2['away_goals'] + mydf2['home_goals']
print(counts)

##### There are 10 times more faceoffs than goals in 1500 games.
a = sns.barplot(x='play_id', y = "event", data = counts)
a.set_title('Count of Events over 1500 games')
a.set_ylabel('Event')
a.set_xlabel('Count')


#sns.pointplot(x='season', y = 'total_goals', data = df2)
#sns.pointplot(x='play_num', y = 'event', data = df2)
  
```
